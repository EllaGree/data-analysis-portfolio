```python
import pandas as pd
import numpy as np

data = {
    'Name':       ['Ama', 'Kofi', None, 'Kwame', 'Efua', 'Ama'],
    'Age':        [24, 31, 27, None, 22, 24],
    'Department': ['Sales', 'tech', 'Sales', 'HR', 'TECH', 'Sales'],
    'Salary':     [3200, 5800, 3900, 4500, None, 3200],
    'Email':      ['ama@gmail.com', 'kofi@gmail.com', 'abena@gmail.com', 
                   'kwame@gmail.com', 'efua@gmail.com', 'ama@gmail.com']
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df[df['Name'].isnull()]
```

```python
df['Name'] = df['Name'].fillna('Unknown')
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df['Department'] = df['Department'].str.title()
df
```

```python
# Check current types
df.dtypes
```

```python
df['Age'] = df['Age'].astype(int)
df.dtypes
```

```python
df
```

```python
df['Department'] = df['Department'].replace('Hr', 'HR')
df
```
