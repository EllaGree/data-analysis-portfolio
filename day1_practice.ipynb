```python
data = {
    'Name':       ['Ama', 'Kofi', 'Abena', 'Kwame', 'Efua'],
    'Age':        [24, 31, 27, 35, 22],
    'Department': ['Sales', 'Tech', 'Sales', 'HR', 'Tech'],
    'Salary':     [3200, 5800, 3900, 4500, 5100],
    'Years':      [1, 6, 3, 8, 2]
}

df = pd.DataFrame(data)
df
```

```python
import pandas as pd
from datetime import datetime
```

```python

# Group by Department and calculate average Years
avg_years_per_dept = df.groupby("Department")["Years"].mean().round(2)

print("Average years worked per department:")
print(avg_years_per_dept)
```
