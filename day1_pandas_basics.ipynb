```python
data = {
    'Name':       ['Ama', 'Kofi', 'Abena', 'Kwame', 'Efua'],
    'Age':        [24, 31, 27, 35, 22],
    'Department': ['Sales', 'Tech', 'Sales', 'HR', 'Tech'],
    'Salary':     [3200, 5800, 3900, 4500, 5100],
    'Years':      [1, 6, 3, 8, 2]
}

df = pd.DataFrame(data)
df
```

```python
df['Monthly_Salary'] = df['Salary'] / 12
df
```

```python
df['Bonus'] = df['Salary'] * 0.10
df
```

```python
df['Level'] = df['Years'].apply(lambda x: 'Senior' if x >= 5 else 'Junior')
df
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
df.groupby('Department')['Salary'].agg(['mean', 'min', 'max', 'count'])
```

```python
# Sort by salary highest to lowest
df.sort_values('Salary', ascending=False)
```

```python
# Sort by department, then salary
df.sort_values(['Department', 'Salary'], ascending=[True, False])
```
