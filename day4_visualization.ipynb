```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Name':       ['Ama', 'Kofi', 'Abena', 'Kwame', 'Efua', 'Yaw', 'Akua'],
    'Department': ['Sales', 'Tech', 'Sales', 'HR', 'Tech', 'Tech', 'HR'],
    'Salary':     [3200, 5800, 3900, 4500, 5100, 6200, 4800],
    'Years':      [1, 6, 3, 8, 2, 9, 5],
    'Age':        [24, 31, 27, 35, 22, 38, 29]
}

df = pd.DataFrame(data)

sns.set_theme(style="whitegrid")
print("Ready!")
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Name', y='Salary', palette='blues_d')
plt.title('Salary by Employee')
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Name', y='Salary', hue='Name', legend=False)
plt.title('Salary by Employee')
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Department', y='Salary', hue='Department', legend=False)
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Years', y='Salary', hue='Department', s=100)
plt.title('Years Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Years', y='Salary', marker='o')
plt.title('Salary Trend by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Salary', 'Years', 'Age']].corr(), 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
