```python
import pandas as pd

# Orders table
orders = pd.DataFrame({
    'order_id':   [1, 2, 3, 4, 5, 6],
    'customer_id':[101, 102, 101, 103, 102, 104],
    'Amount':     [250, 180, 320, 450, 210, 380]
})

# Customers table
customers = pd.DataFrame({
    'customer_id':[101, 102, 103],
    'Name':       ['Ama', 'Kofi', 'Abena'],
    'City':       ['Accra', 'Kumasi', 'Takoradi']
})

# 1. Inner join on customer_id
merged = pd.merge(orders, customers, on='customer_id', how='inner')
print("Inner Join:\n", merged)

# 2. Total amount spent per customer
total_spent = merged.groupby('Name')['Amount'].sum().reset_index()
print("\nTotal Amount Spent per Customer:\n", total_spent)

# 3. City with highest average order amount
avg_city = merged.groupby('City')['Amount'].mean().reset_index()
highest_avg_city = avg_city.loc[avg_city['Amount'].idxmax()]
print("\nCity with Highest Average Order Amount:\n", highest_avg_city)

# 4. Number of orders per customer
order_count = merged.groupby('Name')['order_id'].count().reset_index(name='OrderCount')
print("\nNumber of Orders per Customer:\n", order_count)
```
