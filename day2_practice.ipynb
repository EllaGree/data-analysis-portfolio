```python
import pandas as pd
import numpy as np
data = {
    'Product':  ['Laptop', 'Phone', None, 'Tablet', 'Laptop'],
    'Category': ['ELECTRONICS', 'electronics', 'Electronics', 'ELECTRONICS', 'ELECTRONICS'],
    'Price':    [1500, 800, 1200, None, 1500],
    'Stock':    [10, 25, 30, 15, 10]
}

df = pd.DataFrame(data)
df
```

```python
# Find missing values
missing_values = df.isnull().sum()

print("Missing values per column:")
print(missing_values)

# Optionally, show the actual rows with missing values
print("\nRows with missing values:")
print(df[df.isnull().any(axis=1)])
```

```python
# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()

print("Data without duplicates:")
print(df_no_duplicates)
```

```python
# Fix Category casing (make all uppercase)
df["Category"] = df["Category"].str.upper()

print(df)
```

```python
# Fix Category casing (make all uppercase)
df["Category"] = df["Category"].str.title()

print(df)
```

```python
# Check data types
print(df.dtypes)

# Or get a more detailed summary
print("\nDetailed info:")
print(df.info())
```
