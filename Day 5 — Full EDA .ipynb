```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')
sns.set_theme(style="whitegrid")
df.head()
```

```python
# Shape, columns, data types, missing values
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
avg_tip = df['tip'].mean()
print("Average Tip:", avg_tip)
```

```python
smoker_tips = df.groupby('smoker')['tip'].mean()
print(smoker_tips)
```

```python
day_totals = df.groupby('day')['total_bill'].sum().sort_values(ascending=False)
print(day_totals)
```

```python
corr = df['total_bill'].corr(df['tip'])
print("Correlation between total bill and tip:", corr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Chart 1: Average tip by smoker status
plt.figure(figsize=(6,4))
sns.barplot(x='smoker', y='tip', data=df, estimator='mean', palette='pastel')
plt.title("Average Tip by Smoker Status")
plt.xlabel("Smoker")
plt.ylabel("Average Tip ($)")
plt.show()

# Chart 2: Total bill by day
plt.figure(figsize=(6,4))
sns.barplot(x='day', y='total_bill', data=df, estimator=sum, palette='muted')
plt.title("Total Bill by Day")
plt.xlabel("Day of Week")
plt.ylabel("Total Bill ($)")
plt.show()

# Chart 3: Scatter plot of total bill vs tip
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day', palette='Set2')
plt.title("Relationship Between Total Bill and Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```
