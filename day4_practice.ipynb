```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
pythondata = {
    'Month':   ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales':   [12000, 15000, 13500, 17000, 16000, 19000],
    'Expenses':[8000, 9000, 8500, 10000, 9500, 11000],
    'Profit':  [4000, 6000, 5000, 7000, 6500, 8000]
}

df = pd.DataFrame(pythondata)

# 1. Bar chart — Sales by Month
plt.figure(figsize=(6,4))
plt.bar(df['Month'], df['Sales'], color='skyblue')
plt.title("Sales by Month")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.show()

# 2. Line chart — Profit trend across months
plt.figure(figsize=(6,4))
plt.plot(df['Month'], df['Profit'], marker='o', color='green')
plt.title("Profit Trend Across Months")
plt.xlabel("Month")
plt.ylabel("Profit Amount")
plt.show()

# 3. Scatter plot — Sales vs Profit relationship
plt.figure(figsize=(6,4))
plt.scatter(df['Sales'], df['Profit'], color='red')
plt.title("Sales vs Profit Relationship")
plt.xlabel("Sales Amount")
plt.ylabel("Profit Amount")
plt.show()
```
