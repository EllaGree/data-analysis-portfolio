```python
import matplotlib.pyplot as plt

# Data
year_sales = {
    2003: 3516979.54,
    2004: 4724162.60,
    2005: 1791486.71
}

product_sales = {
    "Classic Cars": 3919615.66,
    "Vintage Cars": 1903150.84,
    "Motorcycles": 1166388.34,
    "Trucks and Buses": 1127789.84,
    "Planes": 975003.57,
    "Ships": 714437.13,
    "Trains": 226243.47
}

country_sales = {
    "USA": 3627982.83,
    "Spain": 1215686.92,
    "France": 1110916.52,
    "Australia": 630623.10,
    "UK": 478880.46,
    "Italy": 374674.31,
    "Finland": 329581.91,
    "Norway": 307463.70,
    "Singapore": 288488.41,
    "Denmark": 245637.15
}

# Chart 1 â Bar chart: Total Sales by Year
plt.figure(figsize=(6,4))
plt.bar(year_sales.keys(), year_sales.values(), color='skyblue')
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales (USD)")
plt.show()

# Chart 2 â Horizontal bar chart: Revenue by Product Line
plt.figure(figsize=(8,5))
plt.barh(list(product_sales.keys()), list(product_sales.values()), color='lightgreen')
plt.title("Revenue by Product Line")
plt.xlabel("Sales (USD)")
plt.ylabel("Product Line")
plt.show()

# Chart 3 â Bar chart: Top 10 Countries by Revenue
plt.figure(figsize=(10,6))
plt.bar(country_sales.keys(), country_sales.values(), color='salmon')
plt.title("Top 10 Countries by Revenue")
plt.xlabel("Country")
plt.ylabel("Sales (USD)")
plt.xticks(rotation=45)
plt.show()
```
