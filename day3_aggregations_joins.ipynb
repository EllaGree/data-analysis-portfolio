```python
import pandas as pd
import numpy as np

# Employees table
employees = pd.DataFrame({
    'emp_id':     [1, 2, 3, 4, 5],
    'Name':       ['Ama', 'Kofi', 'Abena', 'Kwame', 'Efua'],
    'dept_id':    [101, 102, 101, 103, 102],
    'Salary':     [3200, 5800, 3900, 4500, 5100]
})

# Departments table
departments = pd.DataFrame({
    'dept_id':    [101, 102, 103, 104],
    'Department': ['Sales', 'Tech', 'HR', 'Finance'],
    'Manager':    ['Mensah', 'Asante', 'Boateng', 'Adjei']
})

print("Employees:")
print(employees)
print("\nDepartments:")
print(departments)
```

```python
inner = pd.merge(employees, departments, on='dept_id', how='inner')
inner
```

```python
left = pd.merge(employees, departments, on='dept_id', how='left')
left
```

```python
right = pd.merge(employees, departments, on='dept_id', how='right')
right
```

```python
# Join first, then analyze
df = pd.merge(employees, departments, on='dept_id', how='inner')

# Average salary per department
df.groupby('Department')['Salary'].mean()
```

```python
# Count employees per department
df.groupby('Department')['Name'].count()
```

```python
# Multiple stats per department
df.groupby('Department')['Salary'].agg(['mean', 'min', 'max', 'count'])
```
